## Setup

```python
!pip3 install pandas
!pip3 install tensorflow-datasets

import os, glob
import pandas as pd
from sklearn.model_selection import train_test_split
from google.colab import files
import tensorflow_datasets as tfds

train_set_name = 'train'
validation_set_name = 'validation'
test_set_name = 'test'
datasets_type = '.tsv'
datasets_separator = '\t'

choosen_dataset = train_set_name
```

## Prepare dataset

```python
ds, info = tfds.load('scientific_papers', split=choosen_dataset, shuffle_files=True, with_info=True)
```

# Export dataset

```python
df = tfds.as_dataframe(ds, info)
ds, info = None, None
y = df['abstract']
X = df.drop(['abstract', 'section_names'], axis=1)
```

```python
print("Dataset size: " + str(X.shape[0]))
```

```python
dataset = pd.concat([X, y], axis=1)
dataset.to_csv("/content/drive/MyDrive/dataset/" + choosen_dataset + "_sum" + datasets_type, index=False, sep=datasets_separator)
files.download(choosen_dataset + datasets_type)
```
